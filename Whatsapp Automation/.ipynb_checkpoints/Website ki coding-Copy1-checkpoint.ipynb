```python
var_file_contact = r"C:\Users\Asif.Kasi\Documents\CNCD WORK\UNI-k-projects\Whatsapp Automation\contacts - Copy.xlsx"
df = pd.read_excel(var_file_contact)
df
# seperating rows having null values in {study_IDs,Contact and File Name} columns and deleting from DF

df_errors = pd.DataFrame()

df_errors =df_errors.append(df[~df.study_id.notnull()])
df_errors =df_errors.append(df[~df.Contact.notnull()])
df_errors =df_errors.append(df[~df["File Name"].notnull()])
df_errors = df_errors.drop_duplicates()
df_errors
df_tem = pd.DataFrame()
df_tem["study_id"] = df_errors["study_id"]
df = (df.merge(df_tem, on='study_id', how='left', indicator=True)
    .query('_merge == "left_only"')
    .drop('_merge', 1))

df = df.reset_index(drop=True)
df
```

```python
str(df["Contact"][0])[:-2]
```

```python
#changing phone number into number with country code
# df = df.astype(str)
list_write_country_code_with_num = [] 

for i in range(len(df)):
#     print(df["Contact"][i])
    list_write_country_code_with_num.append(phonenumbers.format_number(phonenumbers.parse(str(df["Contact"][i])[:-2], "PK"), phonenumbers.PhoneNumberFormat.INTERNATIONAL))
    
df["Contact"] = list_write_country_code_with_num 
df
```
